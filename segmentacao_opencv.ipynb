```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Lendo a imagem
img = cv2.imread("Lenna.png")

#Prepare a imagem com um filtro médio de 11X11
kernel = np.ones((11, 11), np.float32)/121
dst = cv2.filter2D(img, -1, kernel)

# Mudando a imagem do espaço BGR para o espaço RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)

# Mostrando o resultado
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(dst), plt.title('Média')
plt.xticks([]), plt.yticks([])
plt.show()
```

####Usando a biblioteca do OPenCV

```python
# Lendo a imagem
img = cv2.imread("Lenna.png", 0)

# Executando o filtro Laplaciano
laplacian = cv2.Laplacian(img, cv2.CV_64F)

# Encontrando a borda vertical
# Chamando a função do filtro Sobel dentro da biblioteca do OpenCV
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)

# Encontrando a borda horizontal
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

plt.subplot(2,2,1), plt.imshow(img, cmap='gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2), plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3), plt.imshow(sobelx, cmap='gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4), plt.imshow(sobely, cmap='gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
```
